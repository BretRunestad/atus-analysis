```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn
import numpy as np
```

```python
plt.rc('figure', figsize=(16, 6))
```

```python
%matplotlib inline
```

```python
summary = pd.read_csv("atusdata/atussum_2013/atussum_2013.dat")
summary.info()
```

```python
summary.head()
```

```python
summary =summary.replace(-1, np.nan)
```

```python
adults_crit = (summary.TEAGE >= 18)
no_children_crit = (summary.TRCHILDNUM == 0)
men_crit = (summary.TESEX == 1)
```

```python
adults_with_no_children = summary[adults_crit & no_children_crit]
adults_with_children = summary[adults_crit & ~no_children_crit]
men_with_no_children = summary[adults_crit & no_children_crit & men_crit]
men_with_children = summary[adults_crit & ~no_children_crit & men_crit]
women_with_no_children = summary[adults_crit & no_children_crit & ~men_crit]
women_with_children = summary[adults_crit & ~no_children_crit & ~men_crit]
```

```python
def activity_columns(data, activity_code):
    """For the activity code given, return all columns that fall under that activity."""
    col_prefix = "t{}".format(activity_code)
    return [column for column in data.columns if re.match(col_prefix, column)]
```

```python
def average_minutes(data, activity_code):
    cols = activity_columns(data, activity_code)
    activity_data = data[cols]
    activity_sums = activity_data.sum(axis=1)
    data = data[['TUFINLWGT']]
    data['minutes'] = activity_sums
    data = data.rename(columns={"TUFINLWGT": "weight"})
    data['weighted_minutes'] = data.weight * data.minutes
    return data.weighted_minutes.sum() / data.weight.sum()
```

```python
average_minutes(adults_with_children, "02")
```

```python
average_minutes(adults_with_no_children, "02")
```

```python
mwc_all_household = average_minutes(men_with_children, "02")
mwc_housework = average_minutes(men_with_children, "0201")
mwc_food_prep = average_minutes(men_with_children, "0202")
mwc_interior = average_minutes(men_with_children, "0203")
mwc_exterior = average_minutes(men_with_children, "0204")
mwc_financial = average_minutes(men_with_children, "020901")
mwc_house = [mwc_housework, mwc_food_prep, mwc_interior, mwc_exterior, mwc_financial]
mwc_labels = ["Housework", "Food & Drink Prep", "Interior Maintentance", "Exterior Maintenance", "Financial Management"]
def plot_mwc():
    plt.figure(figsize=(8,8))
    plt.title("Average Minutes of Daily Household tasks for Men with Children")
    plt.pie(mwc_house, labels=mwc_labels)
    plt.show()
```

```python
wwc_all_household = average_minutes(women_with_children, "02")
wwc_housework = average_minutes(women_with_children, "0201")
wwc_food_prep = average_minutes(women_with_children, "0202")
wwc_interior = average_minutes(women_with_children, "0203")
wwc_exterior = average_minutes(women_with_children, "0204")
wwc_financial = average_minutes(women_with_children, "020901")
wwc_house = [wwc_housework, wwc_food_prep, wwc_interior, wwc_exterior, wwc_financial]
wwc_labels = ["Housework", "Food & Drink Prep", "Interior Maintentance", "Exterior Maintenance", "Financial Management"]
def plot_wwc():
    plt.figure(figsize=(8,8))
    plt.pie(wwc_house, labels=wwc_labels)
    plt.title("Average Minutes of Daily Household tasks for Women with Children")
    plt.show()
```
